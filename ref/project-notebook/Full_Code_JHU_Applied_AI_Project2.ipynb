# **Problem Statement**

## **Business Context**

The business under consideration is a **card-issuing financial services organization** that provides **credit and debit card payment services** to retail customers. The organization operates a large-scale digital payment platform that supports **in-store, online, and cross-border card transactions** for its customers.

The organization is headquartered in a **major financial hub**, with operations spanning **multiple countries**, enabling customers to transact across different merchants, devices, and geographic locations in real time. Every day, the business processes a high volume of card transactions across diverse channels and merchant categories.

The organization’s primary function is to **authorize card transactions in real time** while ensuring that fraudulent activity is detected early and legitimate customer transactions are approved without unnecessary friction. This requires balancing **fraud loss prevention** with **customer experience and regulatory compliance**.

However, the organization is facing several challenges in its transaction authorization process. Fraudulent activity often appears as **subtle behavioural deviations**, such as rapid transaction bursts, sudden location changes, usage from new devices, or elevated-risk merchants, which are difficult to capture using static rules alone. At the same time, overly strict controls lead to **false declines**, resulting in customer dissatisfaction and increased operational effort for fraud analysts.

The business engaged you as a Data Scientist to design and implement a behaviour-driven credit card fraud detection solution that improves real-time fraud identification while minimizing false declines. The key challenges involve distinguishing between subtle fraudulent behavior and legitimate customer activity, reducing false positives caused by static rules, and ensuring that fraud decisions are explainable and operationally feasible. The expected outcome is to build a scalable and interpretable fraud detection framework using anomaly detection and machine learning models that enables risk-based actions such as transaction approval, step-up verification, or blocking, thereby reducing fraud losses and improving customer experience.

## Utility Functions

Utility functions help modularize repetitive tasks such as plotting and data transformations. By wrapping common operations (like histograms, boxplots, or correlation heatmaps) into reusable functions:

- Code becomes cleaner and more readable

- Reduces duplication and enforces consistency across analysis

```python
def histogram(col, title, xlabel, ylabel):
    # Create a new figure with a specified size (10 inches wide by 4 inches tall)
    plt.figure(figsize=(10, 4))

    # Plot a histogram of the given column with 50 bins and include a KDE (Kernel Density Estimate) curve
    sns.histplot(col, bins=50, kde=True)

    # Set the title of the plot
    plt.title(title)

    # Label the x-axis with the provided xlabel
    plt.xlabel(xlabel)

    # Label the y-axis with the provided ylabel
    plt.ylabel(ylabel)

    # Display the plot
    plt.show()
```

```python
def barchart(title, xlabel, ylabel, stat='count', x=None, y=None, hue=None, data=None, size=(10, 5)):
    # Create a new figure with the given size (default is 10x5 inches)
    plt.figure(figsize=size)

    # Plot a count plot using seaborn; counts the occurrences of each category by default
    sns.countplot(data=data,stat=stat,x=x,y=y,hue=hue)

    # Set the title of the bar chart
    plt.title(title)

    # Label the x-axis
    plt.xlabel(xlabel)

    # Label the y-axis
    plt.ylabel(ylabel)

    plt.xticks(rotation=45, ha='right')

    # Display the
    if hue is not None:
        plt.legend(title=hue)

    plt.tight_layout()
    plt.show()
```

```python
def boxplot(x, y, data, title, xlabel, ylabel):
    # Create a new figure with the specified size (10 inches wide by 5 inches tall)
    plt.figure(figsize=(10, 5))

    # Create a boxplot using seaborn to show the distribution of y for each category in x
    sns.boxplot(x=x, y=y, data=data)

    # Set the title of the plot
    plt.title(title)

    # Set the label for the x-axis
    plt.xlabel(xlabel)

    # Set the label for the y-axis
    plt.ylabel(ylabel)

    # Rotate x-axis tick labels by 45 degrees for better readability, aligned to the right
    plt.xticks(rotation=45, ha='right')

    # Adjust subplot parameters to give the plot more breathing room
    plt.tight_layout()

    # Display the plot
    plt.show()
```

```python
def stacked_barchart(x, y, xlabel, ylabel, title):
    # Create a DataFrame from the x and y input lists or Series
    df = pd.DataFrame({'x': x, 'y': y})

    # Calculate the proportion (normalized count) of each 'y' category within each 'x' group
    # The result is a DataFrame where rows are values of 'x' and columns are categories of 'y'
    prop_df = df.groupby('x')['y'].value_counts(normalize=True).unstack().fillna(0)

    # Define custom colors for the bars in the stacked bar chart
    colors = ['#1f77b4', '#ff7f0e']  # Adjust as needed for more than two categories

    # Determine how many unique 'x' categories exist to set figure width dynamically
    num_categories = prop_df.shape[0]
    width = max(8, num_categories * 0.6)  # Ensures minimum width of 8, scales with data

    # Plot the normalized stacked bar chart
    prop_df.plot(kind='bar', stacked=True, figsize=(width, 5), color=colors)

    # Set the x-axis label
    plt.xlabel(xlabel)

    # Set the y-axis label
    plt.ylabel(ylabel)

    # Set the chart title
    plt.title(title)

    # Add a legend to explain what each color represents
    plt.legend(title='y')

    # Rotate x-axis labels for readability
    plt.xticks(rotation=45, ha='right')

    # Adjust layout to prevent clipping of labels
    plt.tight_layout()

    # Show the final plot
    plt.show()
```

```python
def scatter_plot_with_hue(x, y, hue, xlabel='', ylabel='', title=''):
    # Create a new figure with the specified size (8 inches wide by 6 inches tall)
    plt.figure(figsize=(8, 6))

    # Create a scatter plot with color based on the 'hue' variable
    # Custom color palette is used for binary classes (0 and 1), with some transparency (alpha=0.6)
    sns.scatterplot(x=x, y=y, hue=hue, palette={0: '#1f77b4', 1: '#ff7f0e'}, alpha=0.6)

    # Set the label for the x-axis
    plt.xlabel(xlabel)

    # Set the label for the y-axis
    plt.ylabel(ylabel)

    # Set the title of the scatter plot
    plt.title(title)

    # Add a legend with a title, using the name of the hue Series/column
    plt.legend(title=str(hue.name))

    # Adjust layout to ensure everything fits well without overlap
    plt.tight_layout()

    # Display the plot
    plt.show()
```

```python
def pairplot_with_hue(data, vars_to_plot, hue_col, title=''):
    # Create a pairplot (scatterplot matrix) for the selected variables with color based on hue_col
    # 'diag_kind="kde"' uses kernel density plots on the diagonal instead of histograms
    # A custom color palette is used for binary categories (0 and 1)
    sns.pairplot(
        data[vars_to_plot + [hue_col]],
        hue=hue_col,
        palette={0: '#1f77b4', 1: '#ff7f0e'},
        diag_kind='kde'
    )

    # Add a title above the entire plot using suptitle (slightly above the default position)
    plt.suptitle(title, y=1.02)

    # Show the complete plot
    plt.show()
```

```python
def heatmap_numerical(data, title='Correlation Heatmap of Numerical Features'):
    # Select only the numerical columns from the dataset
    numeric_data = data.select_dtypes(include='number')

    # Compute the Pearson correlation matrix for the numerical data
    corr = numeric_data.corr()

    # Create a new figure with specified size (10 inches wide by 8 inches tall)
    plt.figure(figsize=(10, 8))

    # Create a heatmap to visualize the correlation matrix
    # - annot=True: display the correlation coefficients inside the cells
    # - fmt=".2f": format the numbers to two decimal places
    # - cmap='coolwarm': use a diverging colormap from blue to red
    # - square=True: force each cell to be square
    # - linewidths=0.5: add lines between cells for readability
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5, vmin=-1, vmax=1)

    # Set the title of the heatmap
    plt.title(title)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)

    # Keep y-axis labels horizontal
    plt.yticks(rotation=0)

    # Adjust layout to prevent clipping of tick labels or title
    plt.tight_layout()

    # Display the heatmap
    plt.show()
```

```python
def pr_plot(y_true, scores, label="model"):
    p, r, _ = precision_recall_curve(y_true, scores)
    ap = average_precision_score(y_true, scores)
    plt.figure()
    plt.plot(r, p, lw=2, label=f"{label} (AP={ap:.3f})")
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title(f"PR Curve - {label}")
    plt.legend(); plt.show()
    return ap
```

```python
def best_f1_threshold(y_true, scores):
    p, r, th = precision_recall_curve(y_true, scores)
    f1 = 2 * p[:-1] * r[:-1] / (p[:-1] + r[:-1] + 1e-9)
    i = np.nanargmax(f1)
    return float(th[i]), float(f1[i]), float(p[i]), float(r[i])
```

## Utility Functions

Before moving ahead, we define a function to check the performance of the model using different metrics.

- We will be using metric functions defined in sklearn for accuracy, precision, recall, f1_score
- We will create a function which will print out all the above metrics in one go.

```python
# defining a function to compute different metrics to check performance of a classification model built using sklearn
def model_performance_classification(model, predictors, target):
    """
    Function to compute different metrics to check classification model performance

    model: classifier
    predictors: independent variables
    target: dependent variable
    """

    # predicting using the independent variables
    pred = model.predict(predictors)

    acc = accuracy_score(target, pred)  # to compute Accuracy
    recall = recall_score(target, pred)  # to compute Recall
    precision = precision_score(target, pred)  # to compute Precision
    f1 = f1_score(target, pred)  # to compute F1-score

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1,},
        index=[0],
    )

    return df_perf
```

```python
def plot_confusion_matrix(model, predictors, target):
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    """
    y_pred = model.predict(predictors)
    cm = confusion_matrix(target, y_pred)
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```
